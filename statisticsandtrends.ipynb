```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extracting dataset
df = pd.read_csv(r"D:\Data Science Assignment\Plastic Waste Around the World.csv")
```

```python
df.describe()
```

```python
# Creating Histogram
plt.figure(figsize=(10, 6)) 
plt.hist(df['Total_Plastic_Waste_MT'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution of Total Plastic Waste')
plt.xlabel('Total Plastic Waste (Metric Tonnes)')
plt.ylabel('Frequency')
plt.grid(axis='y')

# Save the figure
plt.tight_layout()  # Optional: Adjust layout to prevent clipping
plt.savefig("Histogram.png", dpi=300)  # Specify dpi for better quality
plt.show()
```

```python
# Sorting the DataFrame by Total_plastic_waste_MT and selecting the top 10 countries with Total_plastic_waste
top_countries = df.nlargest(10, 'Total_Plastic_Waste_MT')

# Bar Chart for Top 10 Countries
plt.figure(figsize=(10, 6))
plt.bar(top_countries['Country'], top_countries['Total_Plastic_Waste_MT'], color='green', alpha=0.7) 
plt.title('Top 10 Countries by Total Plastic Waste')
plt.xlabel('Country')
plt.ylabel('Total Plastic Waste (Metric Tonnes)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.savefig("barchart.png")
plt.show()
```

```python
#Creating Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Total_Plastic_Waste_MT'], df['Recycling_Rate'], color='purple', alpha=0.7)
plt.title('Total Plastic Waste vs. Recycling Rate')
plt.xlabel('Total Plastic Waste (Metric Tonnes)')
plt.ylabel('Recycling Rate')
plt.grid()
plt.savefig("Scatter.png")
plt.show()
```

```python
# Selecting only numeric columns for correlation
numeric_df = df.select_dtypes(include='number')
```

```python
# Calculating the correlation matrix
correlation_matrix = numeric_df.corr()
```

```python
# Creating a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Heatmap')
plt.savefig("Heatmap.png")
plt.show()
```
